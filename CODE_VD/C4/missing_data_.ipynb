```python
import numpy as np 
import pandas as pd 
arr_none = np.array([1,2,None,3])
print("1. Mảng có None: ")
print(arr_none)
```

```python
print("\n2. Lõi khi tính mean() trên mảng object chứa None: ")
try:   
    arr_none.mean()
except TypeError as e:
    print(f"-> lỗi:{e}")
    
```

```python
for dtype in ['object','int']:
    print('dtype=',dtype)
    %timeit np.arange(1E6,dtype=dtype).sum()
    print()
```

```python
arr_nan = np.array([1,np.nan,3,4])
print(f"mảng gốc:{arr_nan}")
```

```python
# tính chất  của nan 
arr_nan.sum(),arr_nan.min(),arr_nan.max()
```

```python
np.nansum(arr_nan),np.nanmin(arr_nan),np.nanmax(arr_nan)
```
