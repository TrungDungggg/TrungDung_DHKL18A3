# Indexn Preservation


```python
import numpy as np 
import pandas as pd
rng = np.random.RandomState(100)
arr1 = rng.randint(0,10,4)
arr2 = rng.randint(0,10,(3,4))
arr1 
```

```python
arr2 
```

```python
# Tạo series từ mảng numpy 
sr = pd.Series(arr1)
# tạo dt frame  từ ârr 2 với tên cột ABCD
df = pd.DataFrame(arr2,columns=['A','B','C','D'])
print("Series gốc\n:",sr)
print("\nDataFrame gốc\n:",df)
```

```python
# áp dụng ufuncs
print("\n Ser sau np.exp:\n",np.exp(sr))
print("\n DATAFRAME sau np.sin:\n",np.sin(df))
```
