# MULTINDEX ( indexing/slicing) (chọc/lọc)


```python
import pandas as pd
# định nghãi Tuple
index_tuples = [
    ('Hà Nội', 2015), ('Hà Nội', 2020),
    ('Nam Định', 2015), ('Nam Định', 2020),
    ('Thanh Hóa', 2015), ('Thanh Hóa', 2020),
    ('Đà Nẵng', 2015), ('Đà Nẵng', 2020),
    ('Hồ Chí Minh', 2015), ('Hồ Chí Minh', 2020)
]
data = [
    7216000, 8246540,
    1850600, 1780330,
    3514200, 3664940,
    1028800, 1169480,
    8146300, 9227600
]
```

```python
vn_pop = pd.Series(data,index=index_tuples)
print("Ser sử dụng index bằng Tuple:\n",vn_pop)
```

```python
# cho biết dân số HN(2015,2020) và ND(2015,2020)
vn_pop[('Hà Nội',2015):('Nam Định',2015)]
```

```python
vn_pop.index
```

```python
vn_pop[[i for i in vn_pop.index if i[1]=2020]]
```

```python
multi_idx = pd.MultiIndex.from_tuples(index_tuples, names=['Tỉnh/TP', 'Năm'])
multi_idx
```

```python
# reindex series theo multidex vừa tạo 
vn_pop_mul = vn_pop.reindex(multi_idx)
vn_pop_mul # đây là 1 series có chỉ mục đa cấp khác với vn_pop ban đầu
```

```python
# lấy toàn bộ số liệu dân số về năm 2020
vn_pop_mul[:,2020]
```

```python
# CHuyển series thành Dataframe vn_pop_df
vn_pop_df = vn_pop_mul.unstack('Năm') #chuyển Series đa cấp thành DataFrame
print("DataFrame dạng rộng (cột là Năm):")
display(vn_pop_df)

print("\nSeries dạng dài (stack lại):")
display(vn_pop_df.stack())
```

```python
vn_pop_df
```

```python
co_viec_2015 = [3752320, 1080750, 1512340, 512340, 4920000]
co_viec_2020 = [4042451, 1011049, 1600000, 600000, 5200000]
vn_pop_df_expanded = pd.DataFrame({
    ('Tổng dân số',2015):vn_pop_df[2015],
    ('Tổng dân số',2020):vn_pop_df[2020],
    ('Có việc',2015):co_viec_2015,
    ('Có việc',2020):co_viec_2020
},index = vn_pop_df.index)
# tạo multiIndex vn_pop_df.expanded.column từ tuple
vn_pop_df_expanded.columns = pd.MultiIndex.from_tuples(vn_pop_df_expanded.columns,names=['Chỉ tiêu','Năm'])
vn_pop_df_expanded
```

```python
df_co_viec = vn_pop_df_expanded['Có việc']/vn_pop_df_expanded['Tổng dân số']
df_co_viec
```

# Các phương thức tạo Multiindex

```python
#a tạo Multindex cho hàng 
import numpy as np
import pandas as pd # Cần import pandas để dùng pd.DataFrame

df = pd.DataFrame(np.random.rand(4,2),
                  index=[['a', 'a', 'b', 'b'], [1,2,1,2]],
                  columns=['data1', 'data2'])
df
```

```python
import numpy as np
import pandas as pd #
idx_arr = [['a', 'a', 'b', 'b'], [1,2,1,2]]
data = np.random.rand(4,3) # Tạo mảng dữ liệu ngẫu nhiên (4 hàng, 3 cột)
# Tạo đối tượng MultiIndex từ mảng idx_arr
multi_index = pd.MultiIndex.from_arrays(idx_arr, names=['Data1', 'Data2']) # Tạo multiIndex từ mảng
df = pd.DataFrame(data, index=multi_index,
                  columns=['Value1', 'Value2', 'Value3'])
df
```

```python
# tạo Multindex từ tuples 
import numpy as np
import pandas as pd
idx_tuples = [('a', 1), ('a', 2), ('b', 1), ('b', 2)]
data = np.random.randn(4, 3) # Tạo mảng dữ liệu ngẫu nhiên (4 hàng, 3 cột)
df = pd.DataFrame(data, index=idx_tuples,
                  columns=['Value1', 'Value2', 'Value3'])
df
```

```python
import numpy as np
import pandas as pd
index = pd.MultiIndex.from_product([[2020, 2021, 2022], [1, 2, 3]], 
                                   names=['Năm', 'Lần thi'])
columns = pd.MultiIndex.from_product([['Bình', 'Lan', 'Vân'],
                                      ['Toán', 'Lý', 'Hóa']],
                                     names=['Học sinh', 'Môn học'])
# Tạo dữ liệu mẫu, là mảng kích thước 9x9 với các giá trị ngẫu nhiên từ 5-10
data = np.random.randint(5, 10, (9, 9))
score = pd.DataFrame(data, index=index, columns=columns)
score
```
