```python
import pandas as pd 
import numpy as np
print("--- 6. Chuẩn bị DataFrame cho Demo ---")
data = {
    'A': [1.0, 2.0, np.nan, 4.0],
    'B': [np.nan, 6.0, 7.0, 8.0],
    'C': [9, 10, 11, 12]
}
df_demo = pd.DataFrame(data)
print(f"DataFrame gốc (df_demo): \n{df_demo}")
```

```python
print("\n--- 7. Demo isnull() và Lọc ---")
mask_na = df_demo.isnull() # Tạo mặt nạ Boolean cho các giá trị NaN
print(f"Kết quả isnull(): \n{mask_na}")
```

```python
# Lọc các HÀNG không có bất kỳ giá trị NA nào (Chỉ giữ lại hàng sạch)
df_sach = df_demo[df_demo.notnull().all(axis=1)]
print(f"\nDataFrame sau khi lọc các hàng sạch: \n{df_sach}")
```
