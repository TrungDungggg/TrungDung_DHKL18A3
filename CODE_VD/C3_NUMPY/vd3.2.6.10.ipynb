```python
import numpy as np
arr = np.array([[1,2,3],[4,5,6]])

print("Số chiều của mảng là:",arr.ndim)
print("Kích thước của mảng là:",arr.shape)
print("Tổng số phần tử trong mảng:",arr.size)
print("Kiểu dữ liệu của các phần tử trong mảng:",arr.dtype)
print("Kích thước theo(byte) của mỗi phần tử trong mảng là:",arr.itemsize)
print("Tổng kích thước theo(byte) của mảng là:",arr.nbytes)
# bổ sung 
print("Strides ( bước nhảy của bộ nhớ):",arr.sum(axis=0))#[1+4,2+5,3+6]
print("Tổng theo trục 1(theo hàng):",arr.sum(axis=1))#[1+2+3,4+5+6]
```
