```python
import numpy as np 
arr1 = np.array([1, 2, 3]) 
arr2 = np.array([4, 5, 6]) 
Tong = np.add(arr1,arr2) 
Hieu = np.subtract(arr1,arr2) 
Tich = np.multiply(arr1,arr2)  
Thuong = np.divide(arr1,arr2) 
print("Tổng của arr1 + arr2 là:",Tong) 
print("Hiệu của arr1 - arr2 là:",Hieu) 
print("Tích của arr1 * arr2 là:",Tich) 
print("Thương của arr1/arr2 là:",Thuong) 
arr_max = np.maximum(arr1,arr2)
print(arr_max)
```
