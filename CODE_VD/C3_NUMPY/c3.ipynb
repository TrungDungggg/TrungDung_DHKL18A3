```python
import numpy as np
import sys #impo thư viện sys để đo kích thước của đối tượng python = sys.getsizeof()
S = list(range(1000))
# đo kích thước các phần tử trogn ds S 
item_S = sys.getsizeof(S)
print("Kích thước của mỗi phần tử  trong list tính theo bytes:",item_S)
# đo sz của toàn bộ danh sách S 
S_size = item_S*len(S)
print("Kích thước của toàn bộ list tính theo bytes:",S_size)
# tạo 1 mảng numpy array với 1000 phần tử 
D = np.arange(1000)
#đo kích thước của mỗi phần tử trong numpy array 
print("Kích thước mỗi phần tử của Numpy theo bytes:",D.itemsize)
# đo kích thước của toàn bộ mảng numpy array
array_size = D.itemsize*D.size
print("KÍch thước của toàn bộ mảng numpy arrray theo bytes:",array_size)
print(f"Tỷ lệ(list/array):{S_size / array_size:.2f}")
```
