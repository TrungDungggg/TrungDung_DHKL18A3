```python
import sqlite3  
conn = sqlite3.connect('mydatabase.db') 
conn.execute("UPDATE users SET Tên = 'Hà Trung Dũng' where id = '1'" )
conn.commit()
print("Số dòng được cập nhật:",conn.total_changes)
cursor = conn.execute("SELECT * FROM users")
for row in cursor:
    print(row)
conn.close()
```

XÓA BẢN GHI

```python
import sqlite3
conn = sqlite3.connect('mydatabase.db')
conn.execute("DELETE from users where id=3")
conn.commit()
print("Tổng số bản ghi đã được xóa:",conn.total_changes)
cursor = conn.execute("SELECT * from users")
for row in cursor:
    print(row)
conn.close()
```
