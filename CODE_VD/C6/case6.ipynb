```python
import time
import pandas as pd 
files = ["sales_jan.csv","sales_feb.csv","sales_mar.csv"]
def read_file():
    start = time.time()
    for f in files:
        time.sleep(0.2)
        df = pd.read_csv(f)
        print(f"Đọc xong {f}: {len(df)} dòng")
    end = time.time()
    print(f"Thời gian đọc tuần tự:{end - start:.4f} giây")
```

```python
read_file()
```

```python
import threading
import time 
import pandas as pd 
result = []
def load_csv_thread(filename):
    time.sleep(0.2)
    df = pd.read_csv(filename)
    result.append((filename,len(df)))
    print(f"Đọc xong {filename}: {len(df)} dòng")
def read_file_multithread():
    start = time.time()
    threads = []
    for f in files:
        t = threading.Thread(target=load_csv_thread,args=(f,))
        threads.append(t)
        t.start()
    for t in threads:
        t.join()
    end = time.time()
    print(f"Tg đọc đa luồng:{end - start:.4f} s")
```

```python
read_file_multithread()
```

```python
# SS trực tiếp 
if __name__ == "__name__":
    print("____ĐỌC TUẦN TỰ____")
    read_file()
    
    print("___ĐỌC ĐA LUỒNG ___")
    read_file_multithread()
```
