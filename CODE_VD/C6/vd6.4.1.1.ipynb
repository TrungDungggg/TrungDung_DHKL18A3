```python
import threading
import time
x = 0
def increment():
    global x
    tmp = x
    time.sleep(0.0000001) 
    x = tmp + 1
def thread_task():
    """Hàm thread_task() thực hiện tác vụ luồng... thực hiện gọi hàm"""

    for _ in range(100000):
        increment()
def main_task():
    global x
    x = 0
    t1 = threading.Thread(target=thread_task)
    t2 = threading.Thread(target=thread_task)
    # 
    t1.start()
    t2.start()
    # 
    t1.join()
    t2.join()
for i in range(10):
    main_task()
    print("Lần chạy", i+1, "Giá trị cuối cùng của x:", x)
```

```python
import threading
import time
balance = 100 
def add_money():
    global balance
    tmp = balance
    time.sleep(0.001)
    balance= tmp + 10 
threads = [threading.Thread(target=add_money) for _ in range(2)]
for t in threads:
    t.start()
for t in threads:
    t.join()
print("Số dư cuối:",balance)
```
