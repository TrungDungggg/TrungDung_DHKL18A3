```python
import numpy as np 
import pandas as pd
mylist = list('abcdefghijklmnopqrstuvwxyz')
myarr = np.arange(26)
mydict = dict(zip(mylist, myarr))
seri = pd.Series(mydict)
# Chuyển đổi và reset index
df = seri.to_frame().reset_index()
# Đặt tên cột
df.columns = ['index_column', 'value_column']
print(df.head())
```
