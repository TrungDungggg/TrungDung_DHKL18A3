```python
import pandas as pd
ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([4, 5, 6, 7, 8])
# từ s1 xóa các mục có mặt ở s2
res1 = pd.Series(list(set(ser1) - set(ser2)))
print("Kết quả câu 1:")
print(res1)
```

```python
#2 lấy tất cả các mục của ser1 và ser2 nhưng kh nằm chung trong cả 2 
import pandas as pd
only_ser1 = ser1[ser1.isin(ser2) == False]
only_ser2 = ser2[ser2.isin(ser1) == False]
res2 = pd.concat([only_ser1, only_ser2])
print(res2)
```
