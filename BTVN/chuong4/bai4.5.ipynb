```python
import numpy as np 
import pandas as pd 

ser = pd.Series(np.random.normal(10, 5, 25))
#1 Min
print(f"Min: {ser.min():.2f}")
#2 Centile thứ 25
print(f"Kết quả 25%: {ser.quantile(0.25):.2f}")
#3 Trung vị (Median)
print(f"Median: {ser.median():.2f}")
#4 Centile thứ 75
print(f"Kết quả 75%: {ser.quantile(0.75):.2f}")
#5 Max
print(f"Max: {ser.max():.2f}")
```
