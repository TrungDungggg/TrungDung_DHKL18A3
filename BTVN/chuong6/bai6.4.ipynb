```python
import threading
ket_qua = 1
lock = threading.Lock() 
def tinh_giai_thua_doan(start, end):
    global ket_qua
    tam = 1
    for i in range(start, end + 1):
        tam *= i
    with lock:
        ket_qua *= tam
n = 10
t1 = threading.Thread(target=tinh_giai_thua_doan, args=(1, 5))
t2 = threading.Thread(target=tinh_giai_thua_doan, args=(6, 10))

t1.start()
t2.start()
t1.join()
t2.join()
print(f"Giai thừa của {n} là: {ket_qua}")
```
