```python
import threading
import time
def tai_tep(ten_tep):
    print(f"Bắt đầu tải: {ten_tep}...")
    #thời gian tải tệp
    time.sleep(2) 
    print(f"Hoàn thành tải: {ten_tep}")
#ds các tệp cần tải 
danh_sach_file = ["phim_hanh_dong.mp4", "tai_lieu_python.pdf", "nhac_lofi.mp3"]
cac_luong = []
# tạo luồng và khởi chạy
for file in danh_sach_file:
    luong = threading.Thread(target=tai_tep, args=(file,))
    cac_luong.append(luong)
    luong.start()
# kết thúc chương trình
for luong in cac_luong:
    luong.join()
print("Đã tải xong tất cả các tệp!")
```
