```python
import threading
import time

def in_so_chan(nguong):
    print("Luồng số chẵn bắt đầu...")
    for i in range(0, nguong + 1, 2):
        print(f"Số chẵn: {i}")
        time.sleep(0.1) 
def in_so_le(nguong):
    print("Luồng số lẻ bắt đầu...")
    for i in range(1, nguong + 1, 2):
        print(f"Số lẻ: {i}")
        time.sleep(0.1)
# Thiết lập ngưỡng
limit = 20
# Tạo luồng
t_chan = threading.Thread(target=in_so_chan, args=(limit,))
t_le = threading.Thread(target=in_so_le, args=(limit,))

t_chan.start()
t_le.start()
t_chan.join()
t_le.join()

print("Chương trình kết thúc")
```
