```python
import threading
def in_so_chan():
    print("Số chẵn: ", [i for i in range(30, 51) if i % 2 == 0])
def in_so_le():
    print("Số lẻ: ", [i for i in range(30, 51) if i % 2 != 0])
# Tạo 2 luồng
t1 = threading.Thread(target=in_so_chan)
t2 = threading.Thread(target=in_so_le)

t1.start()
t2.start()
t1.join()
t2.join()
```
