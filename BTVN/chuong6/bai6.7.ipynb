```python
import threading
import time
from datetime import datetime

class WebThread(threading.Thread):
    def __init__(self, name, delay, count):
        super().__init__()
        self.name = name
        self.delay = delay
        self.count = count
    def run(self):
        print(f"Starting {self.name}")
        for _ in range(self.count):
            time.sleep(self.delay)
            # In định dạng: Tên: Web e
            now = datetime.now().strftime("%a %b %d %H:%M:%S %Y")
            print(f"{self.name}: Web {now}")
        print(f"Exiting {self.name}")

print("Starting Main Thread")

# Khởi tạo các luồng 
t_google = WebThread("Google", 1, 4)
t_yahoo = WebThread("Yahoo", 2, 4)
t_facebook = WebThread("Facebook", 3, 4)

t_google.start()
t_yahoo.start()
t_facebook.start()

t_google.join()
t_yahoo.join()
t_facebook.join()

print("Exiting Main Thread")
```
