```python
import threading
import time
def cong_viec():
    # tên luồng đang chạy 
    ten_luong = threading.current_thread().name
    print(f"Luồng {ten_luong} đang chạy\n")
    time.sleep(1)
    print(f"Luồng {ten_luong} đã hoàn thành\n")
# Tạo danh sách chứa các luồng
danh_sach_luong = []
# Tạo 5 luồng với tên khác nhau
for i in range(5):
    t = threading.Thread(target=cong_viec, name=f"Thread-UNETI-{i+1}")
    danh_sach_luong.append(t)
    t.start()
# Chờ tất cả các luồng kết thúc
for t in danh_sach_luong:
    t.join()
print("Tất cả các luồng đã kết thú")
```
