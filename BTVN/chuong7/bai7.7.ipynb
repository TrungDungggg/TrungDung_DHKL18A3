```python
import tkinter as tk

def tinh_tong():
    try:
        n = int(entry_n.get())
        tong = sum(range(1, n + 1))
        lbl_kq.config(text=f"Tổng (1 + ... + {n}) là: {tong}")
    except ValueError:
        lbl_kq.config(text="Vui lòng nhập số nguyên!")

root = tk.Tk()
root.title("Tính tổng UNETI")
root.geometry("300x200")

tk.Label(root, text="Nhập số N:").pack(pady=5)
entry_n = tk.Entry(root)
entry_n.pack(pady=5)

btn_tinh = tk.Button(root, text="Tính tổng", command=tinh_tong)
btn_tinh.pack(pady=10)

lbl_kq = tk.Label(root, text="", font=("Arial", 10, "bold"))
lbl_kq.pack(pady=5)

root.mainloop()
```
