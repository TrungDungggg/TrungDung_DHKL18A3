```python
import tkinter as tk

def dao_nguoc_chuoi():
    input_str = entry_word.get()
    reversed_str = ""
    # Thuật toán đảo ngược không dùng hàm có sẵn
    for char in input_str:
        reversed_str = char + reversed_str
    
    lbl_result.config(text=reversed_str)

root = tk.Tk()
root.geometry("400x150")

tk.Label(root, text="Enter a word:").grid(row=0, column=0, padx=10, pady=10)
entry_word = tk.Entry(root)
entry_word.grid(row=0, column=1)

lbl_result = tk.Label(root, text="", fg="blue")
lbl_result.grid(row=1, column=1)

btn_validate = tk.Button(root, text="Validate", command=dao_nguoc_chuoi)
btn_validate.grid(row=2, column=1, pady=10)

root.mainloop()
```
