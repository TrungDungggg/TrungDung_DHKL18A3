```python
import tkinter as tk
from tkinter import messagebox

def kiem_tra_du_lieu():
    # Kiểm tra thông tin
    if ent_ten.get() and ent_id.get() and ent_pass.get():
        messagebox.showinfo("Thông báo", "Gửi thông tin thành công!")
    else:
        messagebox.showwarning("Cảnh báo", "Vui lòng điền đầy đủ thông tin!")
root = tk.Tk()
root.title("Quản lý sinh viên")
root.geometry("350x250")
# Tạo các thành phần giao diện
tk.Label(root, text="Tên sinh viên:").pack(pady=5)
ent_ten = tk.Entry(root)
ent_ten.pack()

tk.Label(root, text="ID sinh viên:").pack(pady=5)
ent_id = tk.Entry(root)
ent_id.pack()

tk.Label(root, text="Mật khẩu:").pack(pady=5)
# show='*' để ẩn mật khẩu khi nhập
ent_pass = tk.Entry(root, show='*')
ent_pass.pack()

# Nút gửi 
btn_gui = tk.Button(root, text="Gửi thông tin", command=kiem_tra_du_lieu)
btn_gui.pack(pady=20)

root.mainloop()
```
