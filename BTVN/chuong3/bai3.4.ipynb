```python
import numpy as np 
arr_zeros = np.zeros(10)
print(arr_zeros)
arr_zeros
```

```python
arr_zeros[5] = 1
print("Cập nhật phần tử ở vị trí thứ 5 là 1:",arr_zeros)
```

```python
arr_h = np.arange(10,25,1)
print(arr_h)
arr_nguoc = arr_h[::-1]
print("In danh sách các phần tử theo thứ tự đảo ngược của arr_h:",arr_nguoc)
```

```python
arr_k = np.array([1,2,0,8,2,0,1,3,0,5,0])
arr_1 = arr_k[arr_k != 0] 
print("Tạo arr_1 với các phần tử khác 0 từ arr_k:",arr_1)
```

```python
arr_1
arr_1_new = np.append(arr_1,[10,20])
print("Thêm phần tử có giá trị 10 và 20 ở cuối array:",arr_1_new)
```

```python
arr_1_new 
arr_1_new = np.insert(arr_1_new,5,100)
print("Thêm phần tử có giá trị = 100 vào vị trí có index=5: ",arr_1_new)
```

```python
arr_1_new
arr_delete= np.delete(arr_1_new,[0,1,2])
print("Xóa các phần tử tại vị trí có index = 0,1,2:",arr_delete)
```
