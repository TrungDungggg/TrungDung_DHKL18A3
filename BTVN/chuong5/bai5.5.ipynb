```python
import sqlite3

# 1. Kết nối đến cơ sở dữ liệu
conn = sqlite3.connect('uneti_db.db')
cursor = conn.cursor()

#2 Tạo bảng DiemSinhVien
cursor.execute("""
CREATE TABLE IF NOT EXISTS DiemSinhVien (
    ma_sv TEXT,
    ho_ten TEXT,
    mon_hoc TEXT,
    diem REAL
)
""")
#3 Chèn điểm sinh viên
ds_diem = [
    ('21101', 'Nguyễn Văn A', 'Lập trình Python', 8.5),
    ('21102', 'Trần Thị B', 'Cơ sở dữ liệu', 9.0),
    ('21103', 'Lê Văn C', 'Mạng máy tính', 7.5),
    ('21104', 'Phạm Thị D', 'Lập trình Python', 10.0)
]

#4 Chèn nhiều bản ghi cùng lúc
cursor.executemany("INSERT INTO DiemSinhVien VALUES (?, ?, ?, ?)", ds_diem)

# Lưu thay đổi
conn.commit()

#5 Truy vấn và hiển thị danh sách điểm
print(f"{'Mã SV':<10} {'Họ tên':<20} {'Môn học':<20} {'Điểm':<5}")
print("-" * 55)

cursor.execute("SELECT * FROM DiemSinhVien")
rows = cursor.fetchall()

for row in rows:
    print(f"{row[0]:<10} {row[1]:<20} {row[2]:<20} {row[3]:<5}")

conn.close()
```
