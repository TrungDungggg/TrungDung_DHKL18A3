```python
import sqlite3

conn = sqlite3.connect('uneti_db.db')
cursor = conn.cursor()

# Cập nhật điểm thành 10 cho môn Lập trình Python
mon_hoc = 'Lập trình Python'
diem_moi = 10.0

cursor.execute("UPDATE DiemSinhVien SET diem = ? WHERE mon_hoc = ?", (diem_moi, mon_hoc))
conn.commit()

print(f"Đã cập nhật điểm môn {mon_hoc} thành {diem_moi} cho tất cả sinh viên ")

conn.close()
```
