```python
import sqlite3

#Tạo kết nối đến database 
conn = sqlite3.connect('uneti_db.db')
cursor = conn.cursor()
# Truy vấn lấy phiên bản của SQLite
cursor.execute("SELECT sqlite_version();")
version = cursor.fetchone()
print(f"Phiên bản SQLite hiện tại: {version[0]}")
# Đóng kết nối
conn.close()
```
