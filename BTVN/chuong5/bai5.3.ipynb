```python
import sqlite3

conn = sqlite3.connect('uneti_db.db')
cursor = conn.cursor()

# Tạo bảng SinhVien 
sql_create_table = """
CREATE TABLE IF NOT EXISTS SinhVien (
    id INTEGER PRIMARY KEY,
    ho_ten TEXT NOT NULL,
    lop TEXT
);
"""
cursor.execute(sql_create_table)
print("Đã tạo bảng SinhVien thành công.")

conn.commit()
conn.close()
```
