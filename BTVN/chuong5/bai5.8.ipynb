```python
import sqlite3

conn = sqlite3.connect('uneti_db.db')
cursor = conn.cursor()

# Xóa sinh viên có mã '21103'
ma_can_xoa = '21103'

cursor.execute("DELETE FROM DiemSinhVien WHERE ma_sv = ?", (ma_can_xoa,))
conn.commit()

if cursor.rowcount > 0:
    print(f"Đã xóa thành công sinh viên có mã: {ma_can_xoa}")
else:
    print(f"Không tìm thấy sinh viên có mã: {ma_can_xoa}")

conn.close()
```
